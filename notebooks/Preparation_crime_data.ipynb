```python
def discount_crimes(discount, df):
  df['disc_index'] = 0
  df['disc_index'] = [discount**(relativedelta(max(df['date']), df.loc[x,'date']).years)*x 
                      for x in df['index']]
  return df
```

```python
discount_factor = 0.9
df2['disc_index'] = 0
N = len(df2['crime_idx'])
for i in range(N):
  df2.loc[i, 'disc_index'] = discount_factor**(relativedelta(max(df2['date']), df2.loc[i,'date']).years)*df2.loc[i,'crime_idx']
```
